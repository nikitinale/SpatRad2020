# Analysis of route contamination with $^{137}Cs$ and exposure of workers to radiation

```python
import numpy as np
import pandas as pd
import math
from sklearn.linear_model import LinearRegression
pd.options.mode.chained_assignment = None

data = pd.read_csv('Data_6-7.xlsx - 6-7_5-15.csv')
data.head()
```

```python
# averaging dose rate from 5 dots
# dose rate on 1 m
data['med_1'] = (data.Y_1m_1+data.Y_1m_2 + data.Y_1m_3 + data.Y_1m_4 + data.Y_1m_5)/5
data['med_1_sd'] = (((data.med_1 - data.Y_1m_1)**2 + 
                     (data.med_1 - data.Y_1m_2)**2 + 
                     (data.med_1 - data.Y_1m_3)**2 + 
                     (data.med_1 - data.Y_1m_4)**2 + 
                     (data.med_1 - data.Y_1m_5)**2)/4)**0.5
# dose rate on surface
data['med_003'] = (data.Y_3cm_1+data.Y_3cm_2 + data.Y_3cm_3 + data.Y_3cm_4 + data.Y_3cm_5)/5
data['med_003_sd'] = (((data.med_1 - data.Y_3cm_1)**2 + 
                       (data.med_1 - data.Y_3cm_2)**2 + 
                       (data.med_1 - data.Y_3cm_3)**2 + 
                       (data.med_1 - data.Y_3cm_4)**2 + 
                       (data.med_1 - data.Y_3cm_5)**2)/4)**0.5
# remove points without measuring
datag = data.dropna(subset=['med_1'])
# convert from nSv/h to mkSv/h
datag[['med_1', 'med_1_sd', 'med_003', 'med_003_sd']] = datag[['med_1', 'med_1_sd', 'med_003', 'med_003_sd']]/1000
datag.loc[:, ['Point', 'med_1', 'med_1_sd', 'med_003', 'med_003_sd']].round(3)
```

```python
# calculating surface density contamination from activity concentration of the radioisotope in soil samples
datag['ПЗ, кБк/м2'] = 0.001*(datag['SA_Soil, Bq/kg']*(datag['M_soil, g']/1000))/((5*math.pi*0.04**2)/4)
datag['ПЗ, Ки/км2'] = 1000**2 * datag['ПЗ, кБк/м2'] / (3.7 * 10**7)
datag['soil_density'] = datag['M_soil, g'] / (5 * 20* (math.pi * 4**2)/4)
datag.loc[~datag['ПЗ, кБк/м2'].isna(), ['Point', 'med_1', 'ПЗ, кБк/м2', 'ПЗ, Ки/км2', 'soil_density']].round(2)
```

```python
import geopandas as gpd
import geoplot as gpl
from matplotlib import pyplot as plt
from shapely.geometry import Point, LineString
import contextily as ctx

# creating geodata
geometry = [Point(xy) for xy in zip(datag.L, datag.B)]
gdata3 = gpd.GeoDataFrame(datag, crs={'init': 'epsg:4326'}, geometry=geometry)
gdata3.head()
```

```python
ax = gdata3.plot(column='med_1', alpha=0.9, figsize=(18,13), 
                legend=True, linewidth=10, cmap='Blues', vmax=0.7, 
                legend_kwds={'orientation': "vertical"})

# creating labels for corner points
tgdata = data.dropna(subset=['UG'])
geom0 = [Point(xy) for xy in zip(tgdata.L, tgdata.B)]
tgdata = gpd.GeoDataFrame(tgdata, crs={'init': 'epsg:4326'}, geometry=geom0)
for idx, row in tgdata.iterrows():
  if (row['UG'][:2] == 'Уг') or (row['UG'][:3] == 'скз'):
    ax.text(row.geometry.x, row.geometry.y, s=row['UG'], fontsize=12)

# creating lines between corner points
trace = tgdata[tgdata['UG'].apply(lambda x: x[:2]) == 'Уг']['geometry'].array
trace = LineString(trace)
trace = gpd.GeoDataFrame(['trace'], crs={'init': 'epsg:4326'}, geometry=[trace])
gtrace = gdata3[gdata3['Point'] > 0]
gtrace['B_b'] = None
gtrace['B_b'][1:] = gtrace['B'][:-1]
gtrace['L_b'] = None
gtrace['L_b'][1:] = gtrace['L'][:-1]
# the variable will contain avarage dose rate in segment between two points
gtrace['med_max'] = None
gtrace['med_max'][1:] = gtrace['med_1'][:-1]
gtrace = gtrace[1:]
gtrace['med_max'] = gtrace.apply(lambda x: np.mean([x.med_max, x.med_1]), axis=1)
geometry2 = [LineString([xy[:2],xy[2:]]) for xy in zip(gtrace.L, gtrace.B, gtrace.L_b, gtrace.B_b)]

# geodata with average dose rates on segments
hazard_trace = gpd.GeoDataFrame(gtrace, crs={'init': 'epsg:4326'}, geometry=geometry2)
trace.plot(ax=ax)
# plot segments with doserate > 0.25 mkSv/h
hazard_trace25 = hazard_trace[gtrace['med_max'] > 0.25]
hazard_trace25.plot(ax=ax, color='yellow', linewidth=5, alpha=0.95, label=">0,25 ${\mu}Sv{\cdot}h^{-1}$")
hazard_trace25['geometry'] = hazard_trace25['geometry'].to_crs(epsg=28406)
# plot segments with doserate > 0.35 mkSv/h
hazard_trace35 = hazard_trace[gtrace['med_max'] > 0.35]
hazard_trace35.plot(ax=ax, color='lightsalmon', linewidth=5, alpha=0.95, label=">0,35 ${\mu}Sv{\cdot}h^{-1}$")
hazard_trace35['geometry'] = hazard_trace35['geometry'].to_crs(epsg=28406)
# plot segments with doserate > 0.45 mkSv/h
hazard_trace45 = hazard_trace[gtrace['med_max'] > 0.45]
hazard_trace45.plot(ax=ax, color='red', linewidth=5, alpha=0.95, label=">0,45 ${\mu}Sv{\cdot}h^{-1}$")
hazard_trace45['geometry'] = hazard_trace45['geometry'].to_crs(epsg=28406)
plt.legend(fontsize=14, title="Hazard zones")
lenghts = {'> 0.25': hazard_trace25['geometry'].length.sum().round(), 
           '> 0.35': hazard_trace35['geometry'].length.sum().round(), 
           '> 0.45': hazard_trace45['geometry'].length.sum().round()}
ctx.add_basemap(ax, crs=gdata3.crs, zoom=15, url=ctx.sources.OSM_A)
plt.title('Gamma dose rate on the route (${\mu}Sv{\cdot}h^{-1}$ and \
           \nzones with the differend radiation hazard level)', 
           fontsize=20)
plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7), sharey=True)
ax1.hist(datag['med_1'], bins=6)
ax1.set_xlabel('${\mu}Sv{\cdot}h^{-1}$', fontsize=14)
ax1.set_ylabel('N', fontsize=14)
ax2.hist(datag['med_003'], bins=6)
ax2.set_xlabel('${\mu}Sv{\cdot}h^{-1}$', fontsize=14)
ax1.set_title('1 m level', fontsize=14)
ax2.set_title('Ground level', fontsize=14)
fig.suptitle("Gamma dose rate distribution on route", fontsize=16)
plt.show()
```

```python
import seaborn as sns
sns.regplot(x='med_1', y='med_003', data=datag)
plt.title('Regression between gamma dose rates\non the ground level and at 1 m height', fontsize=13)
plt.xlabel('1 m level')
plt.ylabel('Ground level')
plt.show()
```

```python
# filter only points where soil samples were took
gdata3c = gdata3.dropna(subset=['ПЗ, Ки/км2'])
gdata3c['ПЗ, Ки/км2'] = gdata3c['ПЗ, Ки/км2'].astype("float32")
gtrace3c = gdata3c[gdata3c['Point'] > 0]

# create segments between the points
# and associate them with average density of contamination with Cs-137
gtrace3c['B_b'] = None
gtrace3c['B_b'][1:] = gtrace3c['B'][:-1]
gtrace3c['L_b'] = None
gtrace3c['L_b'][1:] = gtrace3c['L'][:-1]
gtrace3c['pz_max'] = None
gtrace3c['pz_max'][1:] = gtrace3c['ПЗ, Ки/км2'][:-1]
gtrace3c = gtrace3c[1:]
gtrace3c['pz_max'] = gtrace3c.apply(lambda x: np.mean([x['pz_max'], x['ПЗ, Ки/км2']]), axis=1)
geometry3 = [LineString([xy[:2],xy[2:]]) for xy in zip(gtrace3c.L, gtrace3c.B, gtrace3c.L_b, gtrace3c.B_b)]

gtrace3c = gpd.GeoDataFrame(gtrace3c, crs={'init': 'epsg:4326'}, geometry=geometry3)

# plot density of contamination with Cs-137 on the map
ax = gdata3c.plot(column='ПЗ, Ки/км2', alpha=1, figsize=(18,13), 
                  legend=True, linewidth=5, cmap='OrRd', 
                  vmax=20, legend_kwds={'orientation': "vertical"})

# labels for corners
for idx, row in tgdata.iterrows():
  if (row['UG'][:2] == 'Уг') or (row['UG'][:3] == 'скз'):
    ax.text(row.geometry.x, row.geometry.y, s=row['UG'], fontsize=12)
# put baseman from Open Street Map
gtrace3c.plot(ax=ax, column='pz_max', linewidth=10, alpha=0.9, cmap='OrRd', vmax=20, legend_kwds={'orientation': "horizontal"})
ctx.add_basemap(ax, crs=gdata3.crs, zoom=12, url=ctx.sources.OSM_A)
plt.title('Density of contamination with $^{137}Cs$ of the soil on route ($Ci{\cdot}km^{-2}$)', 
           fontsize=20)
plt.tight_layout()
plt.show()
```

```python
sns.regplot(x='SA_Soil, Bq/kg', y='med_1', data=datag.loc[~datag['SA_Soil, Bq/kg'].isna()])
plt.title('Dependence of gamma dose rate (1 m height) \nfrom the activity concentration of $^{137}Cs$ in the soil', fontsize=13)
plt.xlabel('$Bq{\cdot}kg^{-1}$')
plt.ylabel('${\mu}Sv{\cdot}h^{-1}$')
plt.show()
```

```python
sns.regplot(x='ПЗ, кБк/м2', y='med_1', data=datag.loc[~datag['ПЗ, кБк/м2'].isna()])
plt.title('Dependence of gamma dose rate (1 m height) \nfrom the density of soil contamination with $^{137}Cs$', fontsize=13)
plt.xlabel('$kBq{\cdot}m^{-2}$')
plt.ylabel('${\mu}Sv{\cdot}h^{-1}$')
plt.show()
```

```python
Y = datag.loc[~datag['ПЗ, кБк/м2'].isna(), 'med_1'].values.reshape(-1, 1)
X = datag.loc[~datag['ПЗ, кБк/м2'].isna(), ['ПЗ, кБк/м2']].values.reshape(-1, 1)
X = np.hstack((X, X**2))

regr0 = LinearRegression()
regr0.fit(X, Y)
print(f'R2 = {regr0.score(X, Y)}')
```

```python
Y = datag.loc[~datag['ПЗ, кБк/м2'].isna(), 'med_1'].values.reshape(-1, 1)
X = datag.loc[~datag['ПЗ, кБк/м2'].isna(), ['ПЗ, кБк/м2', 'soil_density']].values.reshape(-1, 2)
X = np.hstack((X, X**2))

regr = LinearRegression()
regr.fit(X, Y)
print(f'R2 = {regr.score(X, Y)}')
```

Soil density almost does not improve model for prediction gamma dose rate from density of contamination

```python
Z = datag.loc[~datag['ПЗ, кБк/м2'].isna(), ['SA_Soil, Bq/kg', 'soil_density']].values.reshape(-1, 2)
Z = np.hstack((Z, Z**2))

regr2 = LinearRegression()
regr2.fit(Z, Y)
print(f'R2 = {regr2.score(Z, Y)}')
```

Activity concentration of $^{137}Cs$ in soil has a slightly less predictive ability for gamma dose rate compare with density contamination. Soil moisture and density of plant cover are able probably to improve the model.
